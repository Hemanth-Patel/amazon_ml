```python
import cv2
import numpy as np
import pytesseract
from matplotlib import pyplot as plt

# Load the input image
image = cv2.imread('/home/drobot/Documents/hemanth/amazon ps/71lTOTqQwtL.jpg')
# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
orig = image.copy()
(H, W) = image.shape[:2]
```

```python
plt.imshow(image_rgb)
```

```python
crop_img = image_rgb[:400,:900]
plt.imshow(crop_img)
```

```python
gray = cv2.cvtColor(np.array(crop_img), cv2.COLOR_RGB2GRAY)
# Apply thresholding (binarization)
_, binary_image = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

plt.imshow(binary_image)
```

```python
plt.imshow(gray)
```

```python
# Use pytesseract to do OCR on the image
text = pytesseract.image_to_string(binary_image)

print(text)
```
