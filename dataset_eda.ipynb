```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/home/drobot/Documents/hemanth/amazon ps/66e31d6ee96cd_student_resource_3/student_resource 3/dataset/train.csv')
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Basic information about the dataset
df.info()
```

```python
# Summary statistics of numerical columns
df.describe()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Check unique values in categorical columns
print("\nUnique values in 'entity_name':", df['entity_name'].unique())
```

```python
# Data distribution of 'entity_value' by 'entity_name'
plt.figure(figsize=(10,6))
sns.boxplot(x='entity_name', y='entity_value', data=df)
plt.title('Entity Value Distribution by Entity Name')
plt.xticks(rotation=45)
plt.show()
```

```python
# Countplot of 'entity_name' to check occurrences
plt.figure(figsize=(10,6))
sns.countplot(x='entity_name', data=df)
plt.title('Count of Each Entity Name')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group the data by 'group_id' and 'entity_name' and analyze the mean of 'entity_value'
grouped = df.groupby(['group_id', 'entity_name'])
grouped.head()
```

```python
# Group the data by 'group_id' and 'entity_name' and analyze the mean of 'entity_value'
grouped = df.groupby(['group_id', 'entity_name'])['entity_value'].mean().unstack()
print(grouped)
```

```python
# Heatmap to visualize the correlation of 'entity_value' across different 'group_id' and 'entity_name'
plt.figure(figsize=(10,6))
sns.heatmap(grouped, annot=True, cmap='coolwarm', fmt=".1f")
plt.title('Mean Entity Value by Group ID and Entity Name')
plt.show()
```
